```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

def f(x):
    return x**4 + 7*x**3 + 5*x**2 - 17*x + 3

def grad_f(x):
  return 4*x**3 + 21*x**2 + 10*x - 17

def g_grafico(path):

  x_values = np.linspace(-6, 2, 1000)
  y_values = f(x_values)

  plt.figure(figsize=(6, 4))
  plt.plot(x_values, y_values, color='blue')

  for _x in path:
    y = f(_x)
    plt.scatter(_x, y, color='red', s=50, zorder=5)

  plt.xlabel('X', fontsize=14)
  plt.ylabel('Y', fontsize=14)
  plt.grid(True)
  plt.show()

g_grafico([-5.5])
```

```python
from scipy.optimize import fsolve
min_point = fsolve(grad_f, [-6])
min_point
```

```python
# Parâmetros iniciais
x = 2
gama = 0.003  # Taxa de aprendizado
beta = 0.92    # Fator de momentum
epsilon = 0.1
max_iterations = 1000
iterations = 0
momentum = 0

path = []

# Executando o método do gradiente com momentum
while iterations < max_iterations:

    path.append(x)
    g_grafico(path)

    grad = grad_f(x)
    momentum = beta * momentum - gama * grad  # Atualiza o momentum
    new_x = x + momentum  # Atualiza a posição usando momentum

    y = f(x)
    print(f"I: {iterations}")
    print(f"f({x}) = {y}")
    print(f"Grad: {grad}")
    print(f"new_x: {new_x}")
    print(f"abs(grad) < epsilon: {abs(grad)} < {epsilon}")
    print("------------------")

    # Condição de parada
    if abs(grad) < epsilon:
        break

    x = new_x
    iterations += 1
    clear_output(wait=True)

# Resultado
print(f"Minimized at: ({x}) after {iterations} iterations")
```
