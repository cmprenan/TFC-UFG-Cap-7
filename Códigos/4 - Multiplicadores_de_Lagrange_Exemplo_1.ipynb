```python
import sympy as sp

# Definir as variáveis
x, y, lambda_ = sp.symbols('x y lambda')

# Definir a função objetivo
f = x**2 + y**2

# Definir a restrição
g = x + y - 2

# Construir a Lagrangiana
L = f + lambda_ * g

# Calcular as derivadas parciais da Lagrangiana
dL_dx = sp.diff(L, x)
dL_dy = sp.diff(L, y)
dL_dlambda = sp.diff(L, lambda_)

# Resolver o sistema de equações
solutions = sp.solve((dL_dx, dL_dy, dL_dlambda), (x, y, lambda_))

# Extrair as soluções
x_opt, y_opt, lambda_opt = solutions[x], solutions[y], solutions[lambda_]
x_opt, y_opt, lambda_opt
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a função objetivo e a restrição
def f(x, y):
    return x**2 + y**2

def g(x, y):
    return x + y - 2

# Criação do grid para plotar a função
x_vals = np.linspace(-1, 3, 400)
y_vals = np.linspace(3, -1, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Plotar os contornos da função objetivo
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)

# Plotar a restrição
x_constraint = np.linspace(-1, 3, 400)
y_constraint = 2 - x_constraint
plt.plot(x_constraint, y_constraint, 'k--', label='$x + y = 2$')

# Marcar a solução ótima considerando a restrição
plt.plot(x_opt, y_opt, 'r*', label='Ótimo - restrição', markersize=10)

# Marcar a solução ótima global, sem retrições
plt.plot(0, 0, 'ro', label='Ótimo - global')

# Adicionar legendas e título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Otimização com Multiplicadores de Lagrange')
plt.legend()
plt.grid(True)
plt.show()
```
